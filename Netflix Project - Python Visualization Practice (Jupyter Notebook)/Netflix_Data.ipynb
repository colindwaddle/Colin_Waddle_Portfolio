```python

import pandas as pd
import matplotlib.pyplot as plt


netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Determining the most frequent 90's movie duration
```

```python
movies = netflix_df[(netflix_df["release_year"] < 2000) & (netflix_df["release_year"] >= 1990) & (netflix_df["type"] == 'Movie')]
print(movies)

plt.hist(movies['duration'])
plt.xlabel('Minutes')
plt.ylabel('No_of_Movies')
plt.title('Duration of Movies In the 90s')

duration = 100
```

```python
#Count the number of short movies (Less than 90 minutes)
```

```python
short_movie_count = 0
for row in movies['duration']:
    if row < 90:
        short_movie_count += 1
    else:
        short_movie_count += 0

print(short_movie_count)
```

```python
#Count the number of short action movies
```

```python
action = movies[movies['genre'] == 'Action']
short_movie_count = 0
for lab, row in action.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1

print(short_movie_count)
```
